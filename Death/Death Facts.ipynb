# Explanation of Metric
The median death age is the median age of people's deaths in Pittsburgh

# Description of dataset
The dataset has the median ages of death by neighborhood in Pittsburgh. It can also be separated by white and black median death ages, but that wasn't looked at in this study.

# Explanation of code
The code imports pandas and the dataset. It then sorts neighborhoods by death age, then grabs the ones with the highest ages. AFter that it removes unnecessary data, reorders, and prints the data.

```python
import pandas as pd
%matplotlib inline

death = pd.read_csv("deathdata.csv", sep=',')
#Get the rows in order or which has the highest median age at death
sorted_death = death.sort_values(by='TOTAL MD AGE AT DEATH')

numToSearch = 43
numToSearch += 3
top_with_empty = sorted_death.tail(numToSearch)
#Get rid of the useless rows
condition = top_with_empty['TOTAL MD AGE AT DEATH'] > -1 #No one dies at -1
top= top_with_empty[condition].copy()

#Reverse order to make them the highest at the top
top = top[::-1]

#The 5 neighborhoods with the highest median age at death
top
```
