```python
import pandas as pd
%matplotlib inline

death = pd.read_csv("deathdata.csv", sep=',')
#Get the rows in order or which has the highest median age at death
sorted_death = death.sort_values(by='TOTAL MD AGE AT DEATH')
#Contrived, just b/c the last 3 are useless
top_8 = sorted_death.tail(8)
#Get rid of the useless rows
condition = top_8['TOTAL MD AGE AT DEATH'] > -1 #No one dies at -1
top_5 = top_8[condition].copy()

#Reverse order to make them the highest at the top
top_5 = top_5[::-1] #LOOOOOOOOK Maybe comment out, might not need to do this.

#The 5 neighborhoods with the highest median age at death
top_5
```
