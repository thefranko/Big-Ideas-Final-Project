```python
import pandas as pd

# Load the CSV
crime = pd.read_csv('crime.csv')

# Create a column marking felonies
crime['Is_Felony'] = crime['ArrestCharge_Felony_Misdemeanor_Description'].astype(str).str.contains('FELONY', case=False, na=False)

# Filter only felony rows
felonies = crime[crime['Is_Felony']].copy()  # copy() is optional but recommended

# Drop the helper column if you want
felonies = felonies.drop(columns='Is_Felony')

# Print the first 20 felony rows
felonies.head(20)
```
